```python
df_final = pd.merge(df_weather, df_air, on="date", how="outer")

df_final.head()
```

```python
(df_final.isnull().sum()).sort_values(ascending=False)
```

```python
df_final["pm10"] = df_final["pm10"].fillna(df_final["pm10"].median())
df_final["pm2_5"] = df_final["pm2_5"].fillna(df_final["pm2_5"].median())               
df_final["carbon_monoxide"] = df_final["carbon_monoxide"].fillna(df_final["carbon_monoxide"].median())
df_final["nitrogen_dioxide"] = df_final["nitrogen_dioxide"].fillna(df_final["nitrogen_dioxide"].median())  
df_final["ozone"] = df_final["ozone"].fillna(df_final["ozone"].median())
df_final["sulphur_dioxide"] = df_final["sulphur_dioxide"].fillna(df_final["sulphur_dioxide"].median()) 

(df_final.isnull().sum()).sort_values(ascending=False)
```

```python
df_final["date_converted"] = pd.to_datetime(df_final["date"], utc=True, errors="coerce")
df_final["data_iso"] = df_final["date_converted"].dt.strftime("%Y-%m-%dT%H:%M:%S")
df_final = df_final.drop(columns=["date_converted"])
```

```python
df_final = df_final.round(2)
```

```python
df_final.columns = df_final.columns.str.upper()
df_final.to_csv("data-cleaning-output.csv", index=False)
```
